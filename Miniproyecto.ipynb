## Normalizacin de los datos

```python
def normalizar(x,xmax,xmin,ymax,ymin):
  m = (ymax-ymin)/(xmax-xmin)
  b = ymin - m*xmin

  y = m*x + b

  return y
```

```python
max = np.max(data).values
min = np.min(data).values
data2 = normalizar(data,max,min,1,-1)
data2.head(10)
```

## Creacin del modelo MLP

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import plot_model
```

Analizando la correlacin entre las variables se decidi que se quitarian las variables 'Beam-draught ratio', 'Length-beam ratio' y 'Length-displacement ratio'.

```python
target = data2['Residuary resistance per unit weight of displacement'].values#y_train
data3 = data2.drop(['Residuary resistance per unit weight of displacement', 'Beam-draught ratio', 'Length-beam ratio', 'Length-displacement ratio'], axis=1)

print(target.shape)
print(data3.shape)
```

## Separar datos de entrenamiento y validacin

```python
from sklearn.model_selection import train_test_split
seed = 42

x_train,x_test,y_train,y_test = train_test_split(data3, target, test_size=0.2, random_state=seed)

print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
input_dim = x_train.shape[1]
num_clases = 1
lr = 1e-3
EPOCHS = 800
loss = 'mse'#'mean_squared_error'
```

#Creador de la arquitectura del modelo 

## Modelo1

```python
def model():
  model = Sequential()
  model.add(Dense(6, input_dim = input_dim, activation='relu'))
  model.add(Dense(3, activation='relu'))
  model.add(Dense(num_clases, activation='linear'))

  model.summary()

  opt = tf.keras.optimizers.SGD(learning_rate=lr)
  
  model.compile(loss = 'MSE', optimizer = opt)

  return model
```

```python
loss = []
loss_val = []
epochs = 1
```

```python
Modelo1 = model()
```

```python
history = Modelo1.fit(x_train,y_train, validation_data=(x_test,y_test), epochs=1, batch_size=32, verbose=0) 

loss.append(history.history['loss'][0])
loss_val.append(history.history['val_loss'][0])

print('Epoch: ',epochs,'loss:',loss[epochs-1], 'val_loss:',loss_val[epochs-1])

epochs += 1

while loss_val[len(loss_val)-1] > 0.003:
  history = Modelo1.fit(x_train,y_train, validation_data=(x_test,y_test), epochs=1, batch_size=32, verbose=0)
  loss.append(history.history['loss'][0])
  loss_val.append(history.history['val_loss'][0])
  print('Epoch: ',epochs,'loss:',loss[epochs-1], 'val_loss:',loss_val[epochs-1])
  epochs += 1
```

### Validacin

```python
score = Modelo1.evaluate(x_test, y_test, verbose=0)

print(score)
```

### Grfica Epochs vs Loss

```python
plt.figure(figsize=(6,6))
plt.plot(loss)
plt.plot(loss_val)
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend(['Train','Test'])
plt.title('Training')
plt.show()
```

### Prediccin

```python
outputTest1 = Modelo1.predict(x_test)
print(outputTest1.shape)
```

```python
plt.figure()
plt.plot(outputTest1, '*r',y_test, '.g')
```

```python
from sklearn import linear_model

output1 = outputTest1
target1 = y_test.reshape([62,1])

print(output1.shape)
print(target1.shape)

regr1 = linear_model.LinearRegression()
regr1.fit(target1, output1)

coef1 = regr1.coef_

print('Coeficiente: ', coef1[0][0])

plt.figure()
plt.plot(target1, output1,'.-b',label = 'Coeficiente de Regresin')
plt.title('Plot')
plt.legend()
plt.show()
```
